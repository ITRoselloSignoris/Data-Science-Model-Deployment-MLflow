```python
import pandas as pd
from funpymodeling import status
import matplotlib.pyplot as plt
import pickle
```

# 1. Cargar datos

```python
data_train = pd.read_csv("../data/raw_data/train.csv" , sep=",", index_col=0)
data_train
```

```python
data_test = pd.read_csv("../data/raw_data/test.csv" , sep=",", index_col=0)
data_test
```

```python
all_data = pd.concat([data_train,data_test], axis = 0)
```

```python
all_data.head(5)
```

```python
all_data.tail(5)
```

```python
status(all_data)
```

# 2. Preparación de datos

```python
data = all_data[['Age', 'Class', 'Inflight wifi service', 'Ease of Online booking', 'Seat comfort', 'Checkin service', 'satisfaction']].copy()
data
```

```python
status(data)
```

```python
data.rename(
    columns={
        "Inflight wifi service":"Wifi",
        "Ease of Online booking":"Booking",
        "Seat comfort":"Seat",
        "Checkin service":"Checkin"
    },
    inplace=True
)

data
```

```python
class_map = {"neutral or dissatisfied":0,"satisfied":1}
data["satisfaction"]=data["satisfaction"].map(class_map)
```

```python
data_ohe = pd.get_dummies(data)
data_ohe
```

# 3. Modelo de Clasificación

```python
x_data = data_ohe.drop("satisfaction", axis=1)
y_data = data_ohe["satisfaction"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=5000, random_state=19)

rf_model.fit(x_train,y_train)
```

# 4. Save model  and ohe_columns

```python
pickle.dump(x_data.columns,open("ohe_categories.pkl","wb"),protocol=pickle.HIGHEST_PROTOCOL)
```

```python
pickle.dump(rf_model,open("model/rf.pkl","wb"))
```
